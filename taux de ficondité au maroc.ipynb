```python
df.rename(columns={"label":"date"},inplace=True)
```

```python
df["Chômage (% de la main-d'oeuvre masculine)"]=df["Chômage (% de la main-d'oeuvre masculine)"]/100
df["Chômage (% de la main-d'oeuvre féminine)"]=df["Chômage (% de la main-d'oeuvre féminine)"]/100
df["PIB par habitant (usd)"]=df["PIB par habitant (usd)"]*10.05
```

## regression liniere de ChômageM

```python
correlation_coefficient = df['date'].corr(df['Chômage (% de la main-d\'oeuvre masculine)'])
predictor = ['date']
correlation_coefficient
```

```python
train_data = df.dropna(subset=['Chômage (% de la main-d\'oeuvre masculine)'])
test_data = df[df['Chômage (% de la main-d\'oeuvre masculine)'].isnull()]
```

```python
X_train = train_data[predictor]
y_train = train_data['Chômage (% de la main-d\'oeuvre masculine)']
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
X_test = test_data[predictor]
predicted_values = model.predict(X_test)
```

```python
df.loc[df['Chômage (% de la main-d\'oeuvre masculine)'].isnull(), 'Chômage (% de la main-d\'oeuvre masculine)'] = predicted_values
```

```python
df
```

## Mme pour chomageF

```python
train_data = df.dropna(subset=['Chômage (% de la main-d\'oeuvre féminine)'])
test_data = df[df['Chômage (% de la main-d\'oeuvre féminine)'].isnull()]
X_train = train_data[predictor]
y_train = train_data['Chômage (% de la main-d\'oeuvre féminine)']
model = LinearRegression()
model.fit(X_train, y_train)
X_test = test_data[predictor]
predicted_values = model.predict(X_test)
df.loc[df['Chômage (% de la main-d\'oeuvre féminine)'].isnull(), 'Chômage (% de la main-d\'oeuvre féminine)'] = predicted_values
```

## SIMPLE LINEAR REGRESSION 

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[["date"]],df["Taux de ficondité"], test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
pred_si_li=model.predict(X_test)
```

```python
mse1 = mean_squared_error(y_test, pred_si_li)
print(f"Mean Squared Error: {mse1}")
```

```python
r1 = r2_score(y_test,pred_si_li )
r1
```

## POLYNOMIALE REGRESSION

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
```

```python
X=df[["PIB par habitant (dh)","ChômageM","ChômageF","Nai_1000","nbr_age_procéa"]]
Y=df["Taux de ficondité"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X3_train, X3_test, y3_train, y3_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)
```

```python
param_grid = {'n_neighbors': [3, 5, 7, 9, 11,13,15]}
```

```python
knn_regressor = KNeighborsRegressor()
```

```python
grid_search = GridSearchCV(knn_regressor, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X3_train, y3_train)
```

```python
j=grid_search.predict(X3_test)
```

```python
mse4 = mean_squared_error(y3_test, j)
print(f"Mean Squared Error: {mse4}")
```

```python
r2 = r2_score(y3_test, j)
r2
```
